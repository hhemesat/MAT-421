The derivative of a function f(x) a x = a is the slope at that point. To get an approximation of the derivative of f(x), we can use the Taylor Series

Let us consider the function f(x) = sin(x), where the serviative of sin(x) is cos(x)

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline
step = 0.1
x = np.arange(0, 2*np.pi, step) 
y = np.sin(x) 

#forward differences
forward_diff = np.diff(y)/step 
x_diff = x[:-1:] 
#exact solution
exact_solution = np.cos(x_diff) 
plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', \
         label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution, \
         label = 'Exact solution')
plt.legend()
plt.show()

# Compute max error
max_error = max(abs(exact_solution - forward_diff))
print(max_error)
```

The graph shows that the two do not exactly line up due tonumerical error in the calculation of the derivative. 

Let us examine how to mitigate the error due to approximation seen above

```python
step = 1
iterations = 20 
step_size = [] 
max_error = [] 

for i in range(iterations):
    step /= 2 
    step_size.append(step) 
    x = np.arange(0, 2 * np.pi, step) 
    y = np.sin(x) 
    #forward differences
    forward_diff = np.diff(y)/step 
    x_diff = x[:-1] 
    #exact solution
    exact_solution = np.cos(x_diff) 
    
    # Compute max error between 
    max_error.append(\
            max(abs(exact_solution - forward_diff)))
plt.figure(figsize = (12, 8))
plt.loglog(step_size, max_error, 'v')
plt.show()
```

This graph shows that the error is proportionate to the step size (the variable called step)
